## Visualizando proporções
Muitas vezes, queremos mostrar como algum grupo, entidade ou quantidade se divide em partes individuais, em que cada uma representa uma proporção do todo. Os exemplos comuns incluem as proporções de homens e mulheres em um grupo de pessoas, as porcentagens de pessoas que votam em diferentes partidos políticos em uma eleição ou as participações de mercado das empresas. O arquétipo dessa visualização é o gráfico de pizza, onipresente em qualquer apresentação de negócios e muito difamado entre os cientistas de dados. Como veremos, visualizar proporções pode ser desafiador, em particular quando o todo é dividido em muitas partes diferentes ou quando queremos ver mudanças nas proporções ao longo do tempo ou entre as condições. Não existe uma visualização ideal única que sempre funcione. Para ilustrar esse problema, discuto alguns cenários diferentes em que cada um exige um tipo diferente de visualização.

## Gráficos de pizza
Gráficos de pizza são gráficos divididos em setores, cada setor da pizza exibe o tamanho de uma parte da informação relacionada. Gráficos de pizza são normalmente utilizados para exibir os tamanhos relativos das partes de um todo.

Tal tipo de informação também pode ser repassada utilizando um gráfico de barras empilhadas ou um gráfico de barras lado a lado. Muitos autores rejeitam categoricamente os gráficos de pizza e argumentam a favor das barras lado a lado ou empilhadas. Outros defendem o uso de gráficos de pizza em algumas ocasiões. Dependendo dos recursos do conjunto de dados e da história específica que deseja contar, você pode preferir uma ou outra abordagem.

Em geral, os gráficos de pizza funcionam bem quando o objetivo é enfatizar frações simples, como metade, um terço ou um quarto. Eles também funcionam bem quando temos conjuntos de dados muito pequenos. Um único gráfico de pizza parece ótimo, mas uma única coluna de barras empilhadas parece estranha. As barras empilhadas, por outro lado, podem funcionar para comparações lado a lado de várias condições ou em uma série de tempo, e as barras lado a lado são preferidas quando queremos comparar diretamente as frações individuais entre si. Um resumo dos vários prós e contras dos gráficos de pizza, barras empilhadas e barras lado a lado é fornecido na Tabela abaixo.

Uma variação dos gráficos de pizza são os gráficos de rosca. A única diferença é que os gráficos de rosca mostra uma área circular não preenchida no meio do gráfico.

# Visualizando associações entre duas ou mais variáveis quantitativas
Muitos conjuntos de dados contêm duas ou mais variáveis quantitativas, e podemos estar interessados em como essas variáveis se relacionam entre si. Por exemplo, podemos ter um conjunto de dados de medições quantitativas de diferentes animais, como altura, peso, comprimento e demandas diárias de energia dos animais. Para representar graficamente a relação de apenas duas dessas variáveis, por ex. a altura e o peso, normalmente usaremos um gráfico de dispersão. Se quisermos mostrar mais de duas variáveis ao mesmo tempo, podemos optar por um gráfico de bolhas, uma matriz de gráfico de dispersão. Finalmente, para conjuntos de dados com dimensões muito elevadas, pode ser útil realizar a redução de dimensão, por exemplo, na forma de análise de componentes principais.

# Gráficos de dispersão
Os gráficos de dispersão ou gráfico de dispersão são representações gráficas do relacionamento entre duas variáveis numéricas. O gráfico de dispersão utiliza pontos para representar essa relação, cada ponto representa o valor de uma variável no eixo horizontal e o valor de outra variável no eixo vertical.

No gráfico acima, temos um scatter plot (ou gráfico de dispersão) do logaritmo do Produto Interno Bruto (PIB) per capita por Expectativa de Vida. Cada ponto do representa valor do logaritmo do PIB per capita de um país em relação à expectativa de vida do país. Observa-se que quanto maior o PIB per capita, maior a expectativa de vida. Ou seja, existe uma correlação positiva entre essas variáveis (à medida que uma variável aumenta, a outra aumenta).

Acompanhe o vídeo abaixo para aprender como criar gráficos de disperção utilizando a biblioteca seaborn em Python: [link](https://www.youtube.com/watch?time_continue=1&v=4yz4cMXCkuw&feature=emb_logo)

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds = diamonds[
    diamonds.cut.isin(['Premium', 'Good']) &
    diamonds.color.isin(['D', 'F', 'J'])
    ].sample(n=100, random_state=22)
diamonds.shape
```

```python
sns.set_style('dark')
```

```python
sns.scatterplot(diamonds.carat, diamonds.price)
```

```python
sns.scatterplot(x='carat', y='price', data=diamonds)
```

```python
sns.scatterplot(x='x', y='y', data=diamonds)
```

```python
sns.set_style('darkgrid')
```

```python
sns.scatterplot(x='carat', y='price', hue='cut',
               data=diamonds)
```

```python
sns.scatterplot(x='carat', y='price', hue='cut',
               data=diamonds,
               palette=['white', 'purple', 'white', '#55cccc', 'white']               
               )
```

```python
sns.scatterplot(x='carat', y='price', hue='depth',
               data=diamonds
               )
```

```python
sns.scatterplot(x='carat', y='price', style='cut',
               data=diamonds,
                s=100
               )
```

```python
sns.scatterplot(x='carat', y='price', size='cut', 
               data=diamonds
               )
```

```python
sns.scatterplot(x='carat', y='price', size='cut', 
               data=diamonds,
               #sizes=100
               )
```

```python
sns.scatterplot(x='carat', y='price', size='depth', 
               data=diamonds,
               #sizes=100
               )
```

```python
sns.scatterplot(x='carat', y='price',
               data=diamonds,
                hue='cut',
                style='cut',
                s=100
               #sizes=100
               )
```

```python
sns.scatterplot(x='carat', y='price',
               data=diamonds,
                hue='cut',
                style='color',
                s=100
               #sizes=100
               )
```

```python
sns.scatterplot(x='carat', y='price',
               data=diamonds,
                hue='cut',
                style='color',
                s=100,
                alpha=0.6
               #sizes=100
               )
```

```python
sns.scatterplot(x='carat', y='price',
               data=diamonds,
                s=200,
                marker='*',
                edgecolor='black'
               #sizes=100
               )
```
