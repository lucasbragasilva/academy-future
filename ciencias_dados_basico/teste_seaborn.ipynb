## Violinplot

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars. shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4, 6, 8])]
```

```python
sns.violinplot(cars.displacement)
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement)
```

```python
sns.violinplot(x='cylinders', y='displacement',
               data=cars,
               hue='origin')
```

```python
sns.violinplot(x='cylinders', y='displacement',  hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])]
              )
plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement',  hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True
              )
plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement',  hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               inner='quartiles'
              )
plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement',  hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               scale='count',
               scale_hue=False
              )
plt.legend(loc=2)
```

```python
(
    cars[cars.origin.isin(['japan', 'europe'])]
    .groupby('cylinders')
    .origin
    .value_counts()
)
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
              order=['japan', 'europe', 'usa'])
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
              linewidth=3)
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement,
              bw=0.2)
```

## Stripplot

```python
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4, 6, 8])]
```

```python
sns.stripplot(cars.weight)
```

```python
sns.stripplot(cars.weight, cars.origin)
```

```python
sns.stripplot(x='weight', y='origin', hue='cylinders',
             data=cars)
```

```python
sns.stripplot(x='weight', y='origin', hue='cylinders',
             data=cars,
             dodge=True)
```

```python
sns.stripplot(data=cars, x='origin', y='weight')
```

```python
sns.stripplot(cars.weight, cars.origin, jitter=False)
```

```python
sns.stripplot(cars.weight, cars.origin, jitter=0.05)
```

```python
from ipywidgets import interactive

def jitter_widget(jitter=0.1):
    sns.stripplot(cars.weight, cars.origin, jitter=jitter)
```

```python
interactive(jitter_widget, jitter=(0, 0.65, 0.05))
```

```python
sns.stripplot(cars.weight, cars.origin, alpha=0.5) #alpha = transparencia
```

```python
sns.stripplot(cars.weight, cars.origin,
              alpha=1,
             size=3)
```

```python
sns.stripplot(cars.weight, cars.origin,
              linewidth=1)
```

```python
sns.stripplot(cars.weight, cars.origin,
              linewidth=1,
             edgecolor='black')
```

```python
sns.stripplot(cars.weight, cars.origin,
              alpha=0.7,
              size=10,
             jitter=0.25,
              marker='*'
             )
```
